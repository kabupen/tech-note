```python
import numpy as np
import seaborn as sns
```

```python
x = np.arange(0,10,0.001)
```

```python
sns.lineplot(x=x, y=np.sin(x))
sns.lineplot(x=x, y=np.sin(5*x))
sns.lineplot(x=x, y=np.sin(10*x))
sns.lineplot(x=x, y=np.sin(20*x))
sns.lineplot(x=x, y=np.sin(x) + np.sin(5*x) + np.sin(10*x) + np.sin(20*x))
```
