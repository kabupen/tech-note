```python
import torch
from torch.nn import functional as F
import numpy as np
```

```python
y_pred = torch.tensor([1, 0, 0, 0], dtype=torch.float32)
y_true = torch.tensor([1, 0, 0, 0], dtype=torch.float32)
loss_ce = torch.nn.CrossEntropyLoss()
print(loss_ce(y_pred, y_true))


y_pred = torch.tensor([0, 1, 0, 0], dtype=torch.float32)
y_true = torch.tensor([1, 0, 0, 0], dtype=torch.float32)
print(loss_ce(y_pred, y_true))

y_pred = torch.tensor([0, 1, 0, 0], dtype=torch.float32)
y_true = torch.tensor([1, 1, 1, 1], dtype=torch.float32)
print(loss_ce(y_pred, y_true))
```

```python
y_pred = torch.tensor([1, 0, 0, 0], dtype=torch.float32)
y_true = torch.tensor([1, 0, 0, 0], dtype=torch.float32)
loss = torch.nn.CrossEntropyLoss()
print(loss(y_pred, y_true))
```
